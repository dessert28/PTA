```python
# 问题一完整代码：基于样本数据预测博主新增关注数

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression

plt.rcParams['font.sans-serif'] = ['SimHei']

df_sample = pd.read_csv("附件1 (Attachment 1).csv")

# 重命名列
df_sample.columns = ['用户ID', '用户行为', '博主ID', '时间']
df_sample['时间'] = pd.to_datetime(df_sample['时间'])

# 映射行为类型
behavior_map = {1: 'view', 2: 'like', 3: 'comment', 4: 'follow'}
df_sample['行为'] = df_sample['用户行为'].map(behavior_map)
df_sample['日期'] = df_sample['时间'].dt.date

# 每日行为统计
pivot = df_sample.pivot_table(index=['博主ID', '日期'], columns='行为', aggfunc='size', fill_value=0).reset_index()
for col in ['view', 'like', 'comment', 'follow']:
    if col not in pivot.columns:
        pivot[col] = 0

# 构造特征
pivot['exposure'] = pivot['view']
pivot['interaction'] = pivot['like'] + pivot['comment']
pivot['conversion'] = pivot['follow'] / (pivot['interaction'] + 1)

# 聚合博主层级数据（不限制时间）
agg_features_all = pivot.groupby('博主ID').agg({
    'exposure': 'mean',
    'interaction': 'mean',
    'conversion': 'mean',
    'follow': 'sum'
}).rename(columns={'follow': 'label'}).reset_index()

# 建模与预测
X_all = agg_features_all[['exposure', 'interaction', 'conversion']]
y_all = agg_features_all['label']
model_all = LinearRegression().fit(X_all, y_all)
agg_features_all['predicted_follow'] = model_all.predict(X_all)

# 提取预测Top5
top5_all = agg_features_all.sort_values(by='predicted_follow', ascending=False).head(5)

# 可视化图1：预测值分布
fig1, ax1 = plt.subplots(figsize=(8, 6))
sns.histplot(agg_features_all['predicted_follow'], bins=10, kde=True, ax=ax1)
ax1.set_title('预测新增关注数分布（样本数据）')

# 图2：曝光度 vs 实际关注数
fig2, ax2 = plt.subplots(figsize=(8, 6))
sns.scatterplot(data=agg_features_all, x='exposure', y='label', ax=ax2)
ax2.set_title('曝光度 vs 实际新增关注数')

# 图3：互动度 vs 实际关注数
fig3, ax3 = plt.subplots(figsize=(8, 6))
sns.scatterplot(data=agg_features_all, x='interaction', y='label', ax=ax3)
ax3.set_title('互动度 vs 实际新增关注数')

# 图4：转化率 vs 实际关注数
fig4, ax4 = plt.subplots(figsize=(8, 6))
sns.scatterplot(data=agg_features_all, x='conversion', y='label', ax=ax4)
ax4.set_title('转化率 vs 实际新增关注数')

# 图5：Top5博主预测值柱状图
fig5, ax5 = plt.subplots(figsize=(10, 6))
sns.barplot(data=top5_all, x='博主ID', y='predicted_follow', ax=ax5)
ax5.set_title('预测新增关注数最多的5位博主（样本数据）')

top5_all[['博主ID', 'predicted_follow']]
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression

# ---------- 一、读取数据 ----------
df1 = pd.read_csv("附件1 (Attachment 1).csv")
df1.columns = ['用户ID', '用户行为', '博主ID', '时间']
df1['时间'] = pd.to_datetime(df1['时间'])
behavior_map = {1: 'view', 2: 'like', 3: 'comment', 4: 'follow'}
df1['行为'] = df1['用户行为'].map(behavior_map)

df2 = pd.read_csv("附件2 (Attachment 2).csv")
df2.columns = ['用户ID', '用户行为', '博主ID', '时间']
df2['时间'] = pd.to_datetime(df2['时间'])
df2['行为'] = df2['用户行为'].map({1: 'view', 2: 'like', 3: 'comment'})

# ---------- 二、构造训练集 ----------
train = df1.groupby(['用户ID', '博主ID', '行为']).size().unstack(fill_value=0).reset_index()
train['hist_view'] = train.get('view', 0)
train['hist_like'] = train.get('like', 0)
train['hist_comment'] = train.get('comment', 0)
train['hist_active'] = train['hist_view'] + train['hist_like'] + train['hist_comment']
train['label'] = (train.get('follow', 0) > 0).astype(int)
train_X = train[['用户ID', '博主ID', 'hist_view', 'hist_like', 'hist_comment', 'hist_active', 'label']]

# ---------- 三、构造预测集 ----------
predict = df2.groupby(['用户ID', '博主ID', '行为']).size().unstack(fill_value=0).reset_index()
predict['day_view'] = predict.get('view', 0)
predict['day_like'] = predict.get('like', 0)
predict['day_comment'] = predict.get('comment', 0)
predict['day_active'] = predict['day_view'] + predict['day_like'] + predict['day_comment']
predict_X = predict[['用户ID', '博主ID', 'day_view', 'day_like', 'day_comment', 'day_active']]

# ---------- 四、合并历史与当日行为 ----------
X = pd.merge(predict_X, train_X, on=['用户ID', '博主ID'], how='left')
X.fillna({'hist_view': 0, 'hist_like': 0, 'hist_comment': 0, 'hist_active': 0, 'label': 0}, inplace=True)

# ---------- 五、模型训练与预测 ----------
features = ['hist_view', 'hist_like', 'hist_comment', 'hist_active',
            'day_view', 'day_like', 'day_comment', 'day_active']
X_train = X[features]
y_train = X['label']

from xgboost import XGBClassifier
model = XGBClassifier(use_label_encoder=False, eval_metric='logloss')
model.fit(X_train, y_train)


X['prob'] = model.predict_proba(X[features])[:, 1]
X['pred'] = (X['prob'] > 0.5).astype(int)

# ---------- 六、表2结果输出 ----------
target_users = ['U7', 'U6749', 'U5769', 'U14990', 'U52010']
table2 = X[(X['用户ID'].isin(target_users)) & (X['pred'] == 1)][['用户ID', '博主ID']] \
    .groupby('用户ID')['博主ID'].apply(list).reset_index()

# 补全空值用户
for user in target_users:
    if user not in table2['用户ID'].values:
        table2 = pd.concat([table2, pd.DataFrame({'用户ID': [user], '博主ID': [[]]})], ignore_index=True)
table2.columns = ['用户ID', '新关注博主ID']

# ---------- 七、可视化 ----------
plt.figure(figsize=(8, 6))
sns.histplot(X['prob'], bins=30, kde=True)
plt.title('预测新增关注概率分布')
plt.xlabel('关注概率')
plt.ylabel('样本数')
plt.savefig("问题2-图1_概率分布.png")
plt.show()

plt.figure(figsize=(8, 6))
sns.scatterplot(data=X, x='day_active', y='prob', hue='label', alpha=0.6)
plt.title('当日活跃度 vs 关注概率')
plt.xlabel('day_active（观看+点赞+评论）')
plt.ylabel('预测概率')
plt.savefig("问题2-图2_活跃度相关性.png")
plt.show()

plt.figure(figsize=(8, 6))
sns.heatmap(X[features + ['prob']].corr(), annot=True, cmap='Spectral')
plt.title('变量相关性热力图')
plt.savefig("问题2-图3_变量热力图.png")
plt.show()

plt.figure(figsize=(6, 5))
sns.countplot(x='pred', data=X)
plt.xticks([0, 1], ['未关注', '已关注'])
plt.title('模型预测标签分布')
plt.ylabel('样本数')
plt.savefig("问题2-图4_预测标签分布.png")
plt.show()

# ---------- 八、输出横向表2结构（修复）----------
user_list = ['U7', 'U6749', 'U5769', 'U14990', 'U52010']
table2_wide = pd.DataFrame(columns=['用户ID'] + user_list)
row = {'用户ID': '新关注博主ID'}
for user in user_list:
    if user in table2['用户ID'].values:
        row[user] = table2[table2['用户ID'] == user]['新关注博主ID'].values[0]
    else:
        row[user] = []
table2_wide = pd.concat([table2_wide, pd.DataFrame([row])], ignore_index=True)
table2_wide
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns

# ========== 一、分块读取并构建用户特征 ==========
file_path = "附件1 (Attachment 1).csv"  # 修改为本地路径
chunksize = 100_000
target_date = pd.to_datetime("2024-07-21").date()
user_action = {}

for chunk in pd.read_csv(file_path, chunksize=chunksize):
    chunk.columns = ['用户ID', '用户行为', '博主ID', '时间']
    chunk['时间'] = pd.to_datetime(chunk['时间'], errors='coerce')
    chunk['是否目标日'] = chunk['时间'].dt.date == target_date

    for row in chunk.itertuples():
        uid = row.用户ID
        if uid not in user_action:
            user_action[uid] = {
                'view': 0, 'like': 0, 'comment': 0, 'follow': 0,
                'dates': set(), 'bloggers': set(), 'label': 0
            }
        if row.用户行为 == 1:
            user_action[uid]['view'] += 1
        elif row.用户行为 == 2:
            user_action[uid]['like'] += 1
        elif row.用户行为 == 3:
            user_action[uid]['comment'] += 1
        elif row.用户行为 == 4:
            user_action[uid]['follow'] += 1
        user_action[uid]['dates'].add(row.时间.date())
        user_action[uid]['bloggers'].add(row.博主ID)
        if row.是否目标日 and row.用户行为 in [2, 3, 4]:
            user_action[uid]['label'] = 1

# ========== 二、整理为特征表 ==========
records = []
for uid, data in user_action.items():
    records.append({
        '用户ID': uid,
        'view': data['view'],
        'like': data['like'],
        'comment': data['comment'],
        'follow': data['follow'],
        'total_actions': data['view'] + data['like'] + data['comment'] + data['follow'],
        'active_days': len(data['dates']),
        'unique_bloggers': len(data['bloggers']),
        'label': data['label']
    })

user_features = pd.DataFrame(records)

# ========== 三、训练XGBoost模型 ==========
features = ['view', 'like', 'comment', 'follow', 'total_actions', 'active_days', 'unique_bloggers']
X = user_features[features]
y = user_features['label']
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

model = XGBClassifier(use_label_encoder=False, eval_metric='logloss')
model.fit(X_train, y_train)

# ========== 四、预测指定用户是否上线 ==========
target_users = ['U9', 'U22405', 'U16', 'U48420']
target_df = user_features[user_features['用户ID'].isin(target_users)].copy()
target_df['pred'] = model.predict(scaler.transform(target_df[features]))

# ========== 五、统计当日互动记录用于推荐 ==========
selected_chunks = []
for chunk in pd.read_csv(file_path, chunksize=chunksize):
    chunk.columns = ['用户ID', '用户行为', '博主ID', '时间']
    chunk['时间'] = pd.to_datetime(chunk['时间'], errors='coerce')
    chunk = chunk[
        (chunk['用户ID'].isin(target_users)) &
        (chunk['时间'].dt.date == target_date) &
        (chunk['用户行为'].isin([2, 3, 4]))
    ]
    if not chunk.empty:
        selected_chunks.append(chunk)

df_online = pd.concat(selected_chunks, ignore_index=True) if selected_chunks else pd.DataFrame(columns=['用户ID', '博主ID'])

# ========== 六、构建表3预测结果 ==========
result = {'用户ID': [], '博主ID 1': [], '博主ID 2': [], '博主ID 3': []}
for uid in target_users:
    is_online = int(target_df[target_df['用户ID'] == uid]['pred'].values[0])
    if is_online == 0:
        result['用户ID'].append(uid)
        result['博主ID 1'].append('')
        result['博主ID 2'].append('')
        result['博主ID 3'].append('')
    else:
        user_data = df_online[df_online['用户ID'] == uid]
        top_bloggers = user_data.groupby('博主ID').size().sort_values(ascending=False).head(3).index.tolist()
        top_bloggers += [''] * (3 - len(top_bloggers))
        result['用户ID'].append(uid)
        result['博主ID 1'].append(top_bloggers[0])
        result['博主ID 2'].append(top_bloggers[1])
        result['博主ID 3'].append(top_bloggers[2])

df_table3 = pd.DataFrame(result)

# ========== 七、影响力指数可视化图表 ==========
# 添加影响力评分
weights = {'view': 0.05, 'like': 0.15, 'comment': 0.15, 'follow': 0.2,
           'total_actions': 0.15, 'active_days': 0.15, 'unique_bloggers': 0.15}
user_features['影响力评分'] = sum(user_features[f] * w for f, w in weights.items())
user_features['影响力指数'] = 100 * (user_features['影响力评分'] - user_features['影响力评分'].min()) / \
                               (user_features['影响力评分'].max() - user_features['影响力评分'].min())

# 图1：影响力指数分布
plt.figure(figsize=(8, 5))
sns.histplot(user_features['影响力指数'], bins=30, kde=True)
plt.title("影响力指数分布")
plt.xlabel("影响力指数")
plt.ylabel("用户数量")
plt.grid(True)
plt.tight_layout()
plt.show()

# 图2：Top20用户
top_users = user_features.sort_values('影响力指数', ascending=False).head(20)
plt.figure(figsize=(10, 6))
sns.barplot(data=top_users, x='影响力指数', y='用户ID', palette='viridis')
plt.title("影响力前20用户")
plt.xlabel("影响力指数")
plt.ylabel("用户ID")
plt.grid(True)
plt.tight_layout()
plt.show()

# 图3：特征热力图
corr_df = user_features[features + ['影响力评分']].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_df, annot=True, cmap='coolwarm')
plt.title("特征与影响力评分相关性热力图")
plt.tight_layout()
plt.show()

# 图4：行为数 vs 影响力指数
plt.figure(figsize=(8, 6))
sns.scatterplot(data=user_features, x='total_actions', y='影响力指数',
                size='unique_bloggers', hue='follow', palette='cool', legend=False)
plt.title("总行为数 vs 影响力指数")
plt.xlabel("总行为数")
plt.ylabel("影响力指数")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split

# ========== 1. 读取数据 ==========
df = pd.read_csv("附件1 (Attachment 1).csv")
df.columns = ['用户ID', '用户行为', '博主ID', '时间']
df['时间'] = pd.to_datetime(df['时间'], errors='coerce')
df.dropna(subset=['时间'], inplace=True)
df['日期'] = df['时间'].dt.date
df['小时'] = df['时间'].dt.hour
df['时段'] = pd.cut(df['小时'], bins=[0, 6, 12, 18, 24],
                    labels=['凌晨', '上午', '下午', '晚上'], right=False)

# ========== 2. 构建用户历史特征 ==========
user_summary = df.groupby('用户ID').agg({
    '用户行为': 'count',
    '博主ID': pd.Series.nunique,
    '日期': pd.Series.nunique
}).reset_index()
user_summary.columns = ['用户ID', '总行为数', '博主数', '活跃天数']

# ========== 3. 构造上线标签 ==========
target_date = pd.to_datetime("2021-07-12").date()
df['是否目标日'] = (df['日期'] == target_date)
online_users = df[df['是否目标日']]['用户ID'].unique()
user_summary['label'] = user_summary['用户ID'].isin(online_users).astype(int)

# ========== 4. 模型训练 ==========
features = ['总行为数', '博主数', '活跃天数']
scaler = MinMaxScaler()
X = scaler.fit_transform(user_summary[features])
y = user_summary['label']
model = XGBClassifier(use_label_encoder=False, eval_metric='logloss')
model.fit(X, y)

# ========== 5. 指定用户上线预测 ==========
target_users = ['U10', 'U1951', 'U1833', 'U26447']
target_df = user_summary[user_summary['用户ID'].isin(target_users)].copy()
target_df['pred'] = model.predict(scaler.transform(target_df[features]))

# ========== 6. 构造推荐表格 ==========
interactions = df[
    (df['用户ID'].isin(target_users)) &
    (df['日期'] == target_date) &
    (df['用户行为'].isin([2, 3, 4]))  # 点赞、评论、关注
]

result = {'用户ID': [], '博主ID 1': [], '时段1': [], '博主ID 2': [], '时段2': [], '博主ID 3': [], '时段3': []}
for uid in target_users:
    if uid in target_df[target_df['pred'] == 1]['用户ID'].values:
        sub = interactions[interactions['用户ID'] == uid]
        top = sub.groupby(['博主ID', '时段']).size().reset_index(name='互动数')
        top_sorted = top.sort_values('互动数', ascending=False).head(3)
        top_entries = top_sorted[['博主ID', '时段']].values.tolist()
        while len(top_entries) < 3:
            top_entries.append(['', ''])
        result['用户ID'].append(uid)
        for i in range(3):
            result[f'博主ID {i+1}'] = result.get(f'博主ID {i+1}', []) + [top_entries[i][0]]
            result[f'时段{i+1}'] = result.get(f'时段{i+1}', []) + [top_entries[i][1]]
    else:
        result['用户ID'].append(uid)
        for i in range(1, 4):
            result[f'博主ID {i}'] = result.get(f'博主ID {i}', []) + ['']
            result[f'时段{i}'] = result.get(f'时段{i}', []) + ['']

df_table4 = pd.DataFrame(result)
print(df_table4)

# ========== 7. 可视化图表 ==========

# 图1：互动时段分布
plt.figure(figsize=(8, 5))
sns.countplot(data=interactions, x='时段', order=['凌晨', '上午', '下午', '晚上'])
plt.title("互动行为时段分布")
plt.xlabel("时段")
plt.ylabel("互动次数")
plt.tight_layout()
plt.show()

# 图2：指定用户小时分布
plt.figure(figsize=(10, 6))
sns.histplot(data=interactions, x='小时', hue='用户ID', bins=24, multiple='stack')
plt.title("目标日每小时互动频率")
plt.xlabel("小时")
plt.ylabel("互动次数")
plt.tight_layout()
plt.show()

# 图3：用户-博主热力图
if not interactions.empty:
    pivot = interactions.groupby(['用户ID', '博主ID']).size().unstack(fill_value=0)
    plt.figure(figsize=(8, 5))
    sns.heatmap(pivot, annot=True, fmt='d', cmap='YlGnBu')
    plt.title("用户-博主互动热力图")
    plt.tight_layout()
    plt.show()
```
